```python
# blanket import modules
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import pprint
```

```python
ufo_data = pd.read_csv("complete.csv",error_bad_lines=False)
ufo_data
```

```python
ufo_data1 = ufo_data.dropna(axis=0, how='any',inplace=False)
ufo_data1 = ufo_data1[ufo_data1['comments'].str.contains("HOAX")==False]
ufo_data1 = ufo_data1[ufo_data1['duration (hours/min)'].str.contains("unknown|year|#")==False]
ufo_data1[["date","time"]] = ufo_data1["datetime"].str.split(n=1, expand=True)
ufo_data1
```

```python
ufo_data2 = ufo_data1.drop(["comments", "date posted", "duration (seconds)", "duration (hours/min)", "datetime", "time"], axis=1)
ufo_data2 = ufo_data2[ufo_data2['city'].str.contains("\(")==False]
ufo_data2 = ufo_data2[["date","city", "state", "country", "shape", "latitude", "longitude"]]
ufo_data2['date'] = ufo_data2['date'].astype('datetime64')
ufo_data2["latitude"] = ufo_data2["latitude"].astype(float)
ufo_data2
```

```python
ufo_data2.to_csv("final_ufo_data.csv", index=False)
```

```python
duration_values = ufo_data2["duration (seconds)"].value_counts().to_frame()
# pd.set_option('display.max_rows', None)
duration_values.index = duration_values.index.astype(float)
duration_values.sort_index()
```
